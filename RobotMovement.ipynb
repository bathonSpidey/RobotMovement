```python
import pandas as pd
import numpy as np 
```

```python
data=pd.read_csv("Filtered.csv")
data.head()
```

```python
data["Target"]=data["BaseRight"]
data.head()
```

```python
for i in range(len(data["Target"])):
    if data["BaseRight"][i]==1:
        data["Target"][i]=1
    elif data["BaseLeft"][i]==1:
        data["Target"][i]=2
    elif data["ShoulderDown"][i]==1:
        data["Target"][i]=3
    elif data["ShoulderUp"][i]==1:
        data["Target"][i]=4
    elif data["ElbowDown"][i]==1:
        data["Target"][i]=5
    elif data["ElbowUp"][i]==1:
        data["Target"][i]=6
    elif data["HandDown"][i]==1:
        data["Target"][i]=7
    elif data["HandUp"][i]==1:
        data["Target"][i]=8  # hand down
    else:
        data["Target"][i]=0
data.head()
```

```python
data.columns
```

```python
data.rename(columns={'DeltaY.1':'DeltaZ'}, inplace=True)
```

```python
data.columns
```

```python
data.drop(['StepNumber', 'DistanceToTarget', 'DeltaBaseRotation',
       'DeltaShoulderPitch', 'DeltaShoulderRoll', 'DeltaElbowPitch',
       'DeltaElbowRoll', 'DeltaHandPitch', 'DeltaGripperRotation', 'DeltaX',
       'DeltaY', 'DeltaZ', 'Gripped', 'IsGood', 'BaseRight', 'BaseLeft',
       'ShoulderDown', 'ShoulderUp', 'ElbowDown', 'ElbowUp', 'HandDown',
       'HandUp'], axis = 1, inplace = True)
data.head()
```

```python
data['Target'] = data['Target'].shift(-1)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.drop(data.tail(2).index,inplace=True)
data.tail()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
dtree_model = DecisionTreeClassifier(max_depth = 121).fit(X_train, y_train)
dtree_predictions = dtree_model.predict(X_test)
```

```python
dtree_predictions
```

```python
cm = confusion_matrix(y_test, dtree_predictions)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, dtree_predictions, target_names=['Nothing','BaseRight', 'BaseLeft',
       'ShoulderDown', 'ShoulderUp', 'ElbowDown', 'ElbowUp',
       'HandUp','HandDown']))
```
